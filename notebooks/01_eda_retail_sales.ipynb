```python
# ============================================================
# Proyecto: Retail Sales Analysis (EDA) - Data Analyst Junior
# Autor: Armando Guizar Ruiz
# Descripci√≥n:
#   Notebook de An√°lisis Exploratorio de Datos (EDA) sobre ventas retail.
#   Incluye: carga de datos, revisi√≥n de calidad, limpieza b√°sica, KPIs,
#   an√°lisis por categor√≠a/tiempo/demograf√≠a y visualizaciones.
#
# Requisitos:
#   - pandas
#   - numpy
#   - matplotlib
#
# Estructura esperada del repo:
#   retail-sales-analysis-python/
#     ‚îú‚îÄ data/raw/retail_sales_dataset.csv
#     ‚îú‚îÄ notebooks/01_eda_retail_sales.ipynb   (este notebook)
#     ‚îî‚îÄ images/ (se guardar√°n gr√°ficas aqu√≠)
# ============================================================

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------
# 1) Configuraci√≥n general
# ---------------------------
plt.rcParams["figure.figsize"] = (10, 6)

# Ruta del dataset (ajusta si tu archivo se llama diferente)
DATA_PATH = "/content/retail_sales_dataset.csv"


# Carpeta para guardar im√°genes (gr√°ficas)
IMAGES_DIR = "../images"
os.makedirs(IMAGES_DIR, exist_ok=True)

print("‚úÖ Entorno listo.")
print(f"üìÑ Dataset esperado en: {DATA_PATH}")
print(f"üñºÔ∏è  Im√°genes se guardar√°n en: {IMAGES_DIR}")

# ---------------------------
# 2) Carga de datos
# ---------------------------
df = pd.read_csv(DATA_PATH)

print("\n‚úÖ Dataset cargado correctamente.")
print("Shape (filas, columnas):", df.shape)

# Vista r√°pida
display(df.head(10))

# ---------------------------
# 3) Exploraci√≥n inicial (calidad)
# ---------------------------
print("\n--- INFO ---")
df_info = df.info()

print("\n--- NULOS POR COLUMNA ---")
display(df.isna().sum().sort_values(ascending=False))

print("\n--- DESCRIBE (num√©ricas) ---")
display(df.describe())

print("\n--- DESCRIBE (categ√≥ricas) ---")
display(df.describe(include=["object"]))

# ---------------------------
# 4) Limpieza y estandarizaci√≥n
# ---------------------------

# 4.1 Estandarizar nombres de columnas (buenas pr√°cticas)
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# 4.2 Renombrar columnas para consistencia (si aplica)
# Nota: Si tu dataset ya est√° en min√∫sculas, esto no cambia nada.
rename_map = {
    "customer_id": "customer_id",
    "product_category": "product_category",
    "price_per_unit": "price_per_unit",
    "total_amount": "total_amount",
    "transaction_id": "transaction_id"
}
df = df.rename(columns=rename_map)

# 4.3 Convertir Date a datetime
# (Si la columna se llama "date", quedar√° como df["date"])
if "date" in df.columns:
    df["date"] = pd.to_datetime(df["date"], errors="coerce")
else:
    raise KeyError("No se encontr√≥ la columna 'date'. Revisa el nombre en tu CSV.")

# 4.4 Estandarizar texto en columnas categ√≥ricas comunes
if "gender" in df.columns:
    df["gender"] = df["gender"].astype(str).str.strip().str.title()

if "product_category" in df.columns:
    df["product_category"] = df["product_category"].astype(str).str.strip().str.title()

# 4.5 Convertir columnas num√©ricas (por seguridad)
# Si alg√∫n valor viene como texto, lo convertimos.
for col in ["age", "quantity", "price_per_unit", "total_amount"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# 4.6 Duplicados
dup_count = df.duplicated().sum()
print(f"\nüîé Duplicados detectados: {dup_count}")
if dup_count > 0:
    df = df.drop_duplicates().copy()
    print("‚úÖ Duplicados eliminados. Nuevo shape:", df.shape)

# 4.7 Revisi√≥n r√°pida de valores inv√°lidos (ej: quantity <= 0)
invalid_qty = df[df["quantity"] <= 0].shape[0] if "quantity" in df.columns else 0
invalid_amount = df[df["total_amount"] < 0].shape[0] if "total_amount" in df.columns else 0

print(f"üîé Registros con quantity <= 0: {invalid_qty}")
print(f"üîé Registros con total_amount < 0: {invalid_amount}")

# (Opcional) Filtrar valores inv√°lidos si existieran
if "quantity" in df.columns:
    df = df[df["quantity"] > 0].copy()
if "total_amount" in df.columns:
    df = df[df["total_amount"] >= 0].copy()

print("\n‚úÖ Limpieza b√°sica completada.")
display(df.head(5))

# ---------------------------
# 5) Feature engineering m√≠nimo (para EDA)
# ---------------------------
df["month"] = df["date"].dt.to_period("M").astype(str)

# Segmentaci√≥n por edad (ajusta bins si tu dataset tiene otros rangos)
bins = [0, 18, 25, 35, 45, 55, 65, 120]
labels = ["<18", "18-25", "26-35", "36-45", "46-55", "56-65", "65+"]

if "age" in df.columns:
    df["age_group"] = pd.cut(df["age"], bins=bins, labels=labels, right=True)

# ---------------------------
# 6) KPIs principales
# ---------------------------
total_revenue = df["total_amount"].sum()
avg_ticket = df["total_amount"].mean()
avg_quantity = df["quantity"].mean()
unique_customers = df["customer_id"].nunique() if "customer_id" in df.columns else np.nan
transactions = df.shape[0]

print("\n================= KPIs =================")
print(f"üí∞ Total Revenue: {total_revenue:,.2f}")
print(f"üßæ Ticket Promedio: {avg_ticket:,.2f}")
print(f"üì¶ Cantidad Promedio por Transacci√≥n: {avg_quantity:,.2f}")
print(f"üë• Clientes √önicos: {unique_customers}")
print(f"üîÅ Transacciones: {transactions}")
print("=======================================")

# ---------------------------
# 7) An√°lisis por categor√≠a
# ---------------------------
sales_by_category = (
    df.groupby("product_category")["total_amount"]
    .sum()
    .sort_values(ascending=False)
)

display(sales_by_category)

ax = sales_by_category.plot(kind="bar")
plt.title("Ingresos por Categor√≠a de Producto")
plt.xlabel("Categor√≠a")
plt.ylabel("Ingresos")
plt.tight_layout()
cat_path = os.path.join(IMAGES_DIR, "sales_by_category.png")
plt.savefig(cat_path, bbox_inches="tight")
plt.show()
print(f"üñºÔ∏è Guardado: {cat_path}")

# ---------------------------
# 8) Tendencia temporal (ventas por mes)
# ---------------------------
monthly_sales = (
    df.groupby("month")["total_amount"]
    .sum()
)

display(monthly_sales)

ax = monthly_sales.plot()
plt.title("Ventas Mensuales (Ingresos)")
plt.xlabel("Mes")
plt.ylabel("Ingresos")
plt.tight_layout()
month_path = os.path.join(IMAGES_DIR, "monthly_sales.png")
plt.savefig(month_path, bbox_inches="tight")
plt.show()
print(f"üñºÔ∏è Guardado: {month_path}")

# ---------------------------
# 9) An√°lisis por g√©nero
# ---------------------------
if "gender" in df.columns:
    gender_sales = df.groupby("gender")["total_amount"].sum().sort_values(ascending=False)
    display(gender_sales)

    ax = gender_sales.plot(kind="bar")
    plt.title("Ingresos por G√©nero")
    plt.xlabel("G√©nero")
    plt.ylabel("Ingresos")
    plt.tight_layout()
    gender_path = os.path.join(IMAGES_DIR, "sales_by_gender.png")
    plt.savefig(gender_path, bbox_inches="tight")
    plt.show()
    print(f"üñºÔ∏è Guardado: {gender_path}")

# ---------------------------
# 10) An√°lisis por rango de edad
# ---------------------------
if "age_group" in df.columns:
    age_sales = df.groupby("age_group")["total_amount"].sum()
    display(age_sales)

    ax = age_sales.plot(kind="bar")
    plt.title("Ingresos por Rango de Edad")
    plt.xlabel("Rango de edad")
    plt.ylabel("Ingresos")
    plt.tight_layout()
    age_path = os.path.join(IMAGES_DIR, "sales_by_age_group.png")
    plt.savefig(age_path, bbox_inches="tight")
    plt.show()
    print(f"üñºÔ∏è Guardado: {age_path}")

# ---------------------------
# 11) Top 10 clientes por ingresos (si aplica)
# ---------------------------
if "customer_id" in df.columns:
    top_customers = (
        df.groupby("customer_id")["total_amount"]
        .sum()
        .sort_values(ascending=False)
        .head(10)
    )
    display(top_customers)

    ax = top_customers.plot(kind="bar")
    plt.title("Top 10 Clientes por Ingresos")
    plt.xlabel("Customer ID")
    plt.ylabel("Ingresos")
    plt.xticks(rotation=45)
    plt.tight_layout()
    topcust_path = os.path.join(IMAGES_DIR, "top_10_customers.png")
    plt.savefig(topcust_path, bbox_inches="tight")
    plt.show()
    print(f"üñºÔ∏è Guardado: {topcust_path}")

# ---------------------------
# 12) Resumen ejecutivo (insights)
# ---------------------------
top_category = sales_by_category.index[0] if len(sales_by_category) > 0 else "N/A"
top_month = monthly_sales.idxmax() if len(monthly_sales) > 0 else "N/A"

print("\n================ INSIGHTS (Borrador) ================")
print(f"1) La categor√≠a con mayor ingreso es: {top_category}.")
print(f"2) El mes con mayor ingreso es: {top_month}.")
print("3) El ticket promedio sugiere oportunidades para estrategias de cross-selling y bundles.")
print("4) La segmentaci√≥n por edad y g√©nero puede guiar campa√±as personalizadas.")
print("5) Recomendaci√≥n: monitorear KPIs mensuales y fortalecer el surtido en categor√≠as top.")
print("=====================================================")

print("\n‚úÖ Notebook EDA ejecutado. Pr√≥ximo paso: documentar insights en el README del proyecto.")
```
